```python
import numpy as np
```

```python
Avogadro = 6.02214076*1e23
Planck = 6.62607015*1e-34
Boltzmann = 1.380649*1e-23
Pi = 3.14159265359

hbar = Planck/(2*Pi)
```

```python
M = 184*1e-3
rho = 19.3*1e3
```

```python
eFermi = np.power((rho*Avogadro/M)*3*Pi**2, 2/3)*(Avogadro*hbar**2/(2*M))
```

```python
TFermi = eFermi/Boltzmann
```
