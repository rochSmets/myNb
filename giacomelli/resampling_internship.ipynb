```python
# moments calculation functions
import numpy as np
import scipy.stats as st
from scipy.special import gamma
import scipy.integrate as inte
import sympy as sym
import random as rand


class Particle:
  def __init__(weight, position, velocity):
    self.weight = weight

class Population:
  def __init__(listOfParticles)
    self.population = listOfParticles

def zeroOrderMoment(population):
  # population = list[particle]
  return np.sum(population.zeroOrderMoment())

def mom_order0(L):
    # calculates zeroth order moment as observed from set L
    mom0 = 0
    for p in L:
        mom0 = mom0 + p[6]
    return mom0

def mom_order1(L):
    # calculates first order moment as observed from set L
    weights = L[:,6]
    positions = L[:,0:3]
    velicities = L[:,3:6]
    mom1 = np.zeros(3)
    for w, p, v in zip(weights, positions, velocities):

    for p in L:
        mom1[0] = mom1[0] + p[6]*p[3]
        mom1[1] = mom1[1] + p[6]*p[4]
        mom1[2] = mom1[2] + p[6]*p[5]
    return mom1

def mom_order2(L):
    # calculates second order moment as observed from set L
    mom2 = np.zeros((3,3))
    u = mom_order1(L)/mom_order0(L)
    ux, uy, uz = u[0], u[1], u[2]
    for p in L:
        mom2[0,0] = mom2[0,0] + p[6]*(p[3] - ux)*(p[3] - ux)
        mom2[0,1] = mom2[0,1] + p[6]*(p[3] - ux)*(p[4] - uy)
        mom2[0,2] = mom2[0,2] + p[6]*(p[3] - ux)*(p[5] - uz)
        mom2[1,1] = mom2[1,1] + p[6]*(p[4] - uy)*(p[4] - uy)
        mom2[1,2] = mom2[1,2] + p[6]*(p[4] - uy)*(p[5] - uz)
        mom2[2,2] = mom2[2,2] + p[6]*(p[5] - uz)*(p[5] - uz)

    mom2[1,0] = mom2[0,1]
    mom2[2,0] = mom2[0,2]
    mom2[2,1] = mom2[1,2]
    return mom2

def mom_order3(L):
    # calculates third order moment as observed from set L
    mom2 = np.zeros((3,3,3))
    u = mom_order1(L)/mom_order0(L)
    ux, uy, uz = u[0], u[1], u[2]
    for p in L:
        mom2[0,0,0] = mom2[0,0,0] + p[6]*(p[3] - ux)*(p[3] - ux)*(p[3] - ux)
        mom2[1,1,1] = mom2[1,1,1] + p[6]*(p[4] - uy)*(p[4] - uy)*(p[4] - uy)
        mom2[0,0,1] = mom2[0,0,1] + p[6]*(p[3] - ux)*(p[4] - uy)*(p[3] - ux)
        mom2[0,0,2] = mom2[0,0,2] + p[6]*(p[3] - ux)*(p[5] - uz)*(p[3] - ux)
        mom2[0,1,1] = mom2[0,1,1] + p[6]*(p[3] - ux)*(p[4] - uy)*(p[4] - uy)
        mom2[0,1,2] = mom2[0,1,2] + p[6]*(p[4] - uy)*(p[5] - uz)*(p[3] - ux)
        mom2[2,0,2] = mom2[2,0,2] + p[6]*(p[5] - uz)*(p[5] - uz)*(p[3] - ux)
        mom2[1,1,2] = mom2[1,1,2] + p[6]*(p[4] - uy)*(p[4] - uy)*(p[5] - uz)
        mom2[2,1,2] = mom2[2,1,2] + p[6]*(p[5] - uz)*(p[5] - uz)*(p[4] - uy)
        mom2[2,2,2] = mom2[2,2,2] + p[6]*(p[5] - uz)*(p[5] - uz)*(p[5] - uz)

    mom2[0,1,0] = mom2[0,0,1]
    mom2[1,0,0] = mom2[0,0,1]
    mom2[0,2,0] = mom2[0,0,2]
    mom2[2,0,0] = mom2[0,0,2]
    mom2[1,0,1] = mom2[0,1,1]
    mom2[1,1,0] = mom2[0,1,1]
    mom2[1,0,2] = mom2[0,1,2]
    mom2[1,2,0] = mom2[0,1,2]
    mom2[0,2,1] = mom2[0,1,2]
    mom2[2,0,1] = mom2[0,1,2]
    mom2[2,1,0] = mom2[0,1,2]
    mom2[2,2,0] = mom2[2,0,2]
    mom2[0,2,2] = mom2[2,0,2]
    mom2[1,2,1] = mom2[1,1,2]
    mom2[2,1,1] = mom2[1,1,2]
    mom2[2,2,1] = mom2[2,1,2]
    mom2[1,2,2] = mom2[2,1,2]
    return mom2
```

```python
def standard_dev(L):
    # calculate standard deviation of a set L
    av = np.zeros(len(L))+np.mean(L)
    L = np.array(L)
    S_2 = np.sum((L-av)**2)/(len(L) - 1)
    return np.sqrt(S_2)

def moments_calc(cells):
    # calculates moments of order zero to three returning a dictionary with the moments, the mean,
    # the range, the standard deviation and the coefficient of variation for each set in a given set of sets
    moments0 = []
    moments1 = []
    moments2 = []
    moments3 = []

    for c in cells:
        moments0.append(mom_order0(c))
        moments1.append(mom_order1(c))
        moments2.append(mom_order2(c))
        moments3.append(mom_order3(c))

    M0 = {'moments' : moments0,
          'mean' : np.mean(moments0),
          'range' : (min(moments0), max(moments0)),
          'st_dev' : standard_dev(moments0),
          'c_var' : standard_dev(moments0)/np.mean(moments0) * 100
         }

    moments1 = np.array(moments1)
    M1mean = np.array([np.mean(moments1[:,0]), np.mean(moments1[:,1]), np.mean(moments1[:,2])])
    M1st = np.array([standard_dev(moments1[:,0]),standard_dev(moments1[:,1]),standard_dev(moments1[:,2])])

    M1 = {'moments' : moments1,
          'mean' : M1mean,
          'range' : np.array([(min(moments1[:,0]), max(moments1[:,0])),
                             (min(moments1[:,1]), max(moments1[:,1])),
                             (min(moments1[:,2]), max(moments1[:,2]))]),
          'st_dev' : M1st,
          'c_var' :  (M1st/M1mean) * 100
         }

    moments2 = np.array(moments2)
    M2mean = np.array([(np.mean(moments2[:,0,0]),np.mean(moments2[:,0,1]), np.mean(moments2[:,0,2])),
                       (np.mean(moments2[:,1,0]),np.mean(moments2[:,1,1]), np.mean(moments2[:,1,2])),
                       (np.mean(moments2[:,2,0]),np.mean(moments2[:,2,1]), np.mean(moments2[:,2,2]))])
    M2st = np.array([(standard_dev(moments2[:,0,0]),standard_dev(moments2[:,0,1]),standard_dev(moments2[:,0,2])),
                     (standard_dev(moments2[:,1,0]),standard_dev(moments2[:,1,1]),standard_dev(moments2[:,1,2])),
                     (standard_dev(moments2[:,2,0]),standard_dev(moments2[:,2,1]),standard_dev(moments2[:,2,2]))])

    M2 = {'moments' : moments2,
          'mean' : M2mean,
          'range' : np.array([((min(moments2[:,0,0]), max(moments2[:,0,0])),
                             (min(moments2[:,0,1]), max(moments2[:,0,1])),
                             (min(moments2[:,0,2]), max(moments2[:,0,2]))),
                            ((min(moments2[:,1,0]), max(moments2[:,1,0])),
                             (min(moments2[:,1,1]), max(moments2[:,1,1])),
                             (min(moments2[:,1,2]), max(moments2[:,1,2]))),
                            ((min(moments2[:,2,0]), max(moments2[:,2,0])),
                             (min(moments2[:,2,1]), max(moments2[:,2,1])),
                             (min(moments2[:,2,2]), max(moments2[:,2,2])))]),
          'st_dev' : M2st,
          'c_var' :  (M2st/M2mean) * 100
         }

    moments3 = np.array(moments3)

    M3mean = np.array([((np.mean(moments3[:,0,0,0]),np.mean(moments3[:,0,1,0]), np.mean(moments3[:,0,2,0])),
                       (np.mean(moments3[:,1,0,0]),np.mean(moments3[:,1,1,0]), np.mean(moments3[:,1,2,0])),
                       (np.mean(moments3[:,2,0,0]),np.mean(moments3[:,2,1,0]), np.mean(moments3[:,2,2,0]))),
                       ((np.mean(moments3[:,0,0,1]),np.mean(moments3[:,0,1,1]), np.mean(moments3[:,0,2,1])),
                       (np.mean(moments3[:,1,0,1]),np.mean(moments3[:,1,1,1]), np.mean(moments3[:,1,2,1])),
                       (np.mean(moments3[:,2,0,1]),np.mean(moments3[:,2,1,1]), np.mean(moments3[:,2,2,1]))),
                       ((np.mean(moments3[:,0,0,2]),np.mean(moments3[:,0,1,2]), np.mean(moments3[:,0,2,2])),
                       (np.mean(moments3[:,1,0,2]),np.mean(moments3[:,1,1,2]), np.mean(moments3[:,1,2,2])),
                       (np.mean(moments3[:,2,0,2]),np.mean(moments3[:,2,1,2]), np.mean(moments3[:,2,2,2])))])

    M3st = np.array([((standard_dev(moments3[:,0,0,0]),standard_dev(moments3[:,0,1,0]),standard_dev(moments3[:,0,2,0])),
                     (standard_dev(moments3[:,1,0,0]),standard_dev(moments3[:,1,1,0]),standard_dev(moments3[:,1,2,0])),
                     (standard_dev(moments3[:,2,0,0]),standard_dev(moments3[:,2,1,0]),standard_dev(moments3[:,2,2,0]))),
                     ((standard_dev(moments3[:,0,0,1]),standard_dev(moments3[:,0,1,1]),standard_dev(moments3[:,0,2,1])),
                     (standard_dev(moments3[:,1,0,1]),standard_dev(moments3[:,1,1,1]),standard_dev(moments3[:,1,2,1])),
                     (standard_dev(moments3[:,2,0,1]),standard_dev(moments3[:,2,1,1]),standard_dev(moments3[:,2,2,1]))),
                     ((standard_dev(moments3[:,0,0,2]),standard_dev(moments3[:,0,1,2]),standard_dev(moments3[:,0,2,2])),
                     (standard_dev(moments3[:,1,0,2]),standard_dev(moments3[:,1,1,2]),standard_dev(moments3[:,1,2,2])),
                     (standard_dev(moments3[:,2,0,2]),standard_dev(moments3[:,2,1,2]),standard_dev(moments3[:,2,2,2])))])

    m3range = np.array([(((min(moments3[:,0,0,0]), max(moments3[:,0,0,0])),(min(moments3[:,0,1,0]), max(moments3[:,0,1,0])),
                        (min(moments3[:,0,2,0]), max(moments3[:,0,2,0]))),((min(moments3[:,1,0,0]), max(moments3[:,1,0,0])),
                        (min(moments3[:,1,1,0]), max(moments3[:,1,1,0])),(min(moments3[:,1,2,0]), max(moments3[:,1,2,0]))),
                        ((min(moments3[:,2,0,0]), max(moments3[:,2,0,0])),(min(moments3[:,2,1,0]), max(moments3[:,2,1,0])),
                        (min(moments3[:,2,2,0]), max(moments3[:,2,2,0])))),(((min(moments3[:,0,0,1]), max(moments3[:,0,0,1])),(min(moments3[:,0,1,0]), max(moments3[:,0,1,0])),
                        (min(moments3[:,0,2,1]), max(moments3[:,0,2,1]))),((min(moments3[:,1,0,1]), max(moments3[:,1,0,1])),
                        (min(moments3[:,1,1,1]), max(moments3[:,1,1,1])),(min(moments3[:,1,2,1]), max(moments3[:,1,2,1]))),
                        ((min(moments3[:,2,0,1]), max(moments3[:,2,0,1])),(min(moments3[:,2,1,1]), max(moments3[:,2,1,1])),
                        (min(moments3[:,2,2,1]), max(moments3[:,2,2,1])))),(((min(moments3[:,0,0,1]), max(moments3[:,0,0,1])),(min(moments3[:,0,1,0]), max(moments3[:,0,1,0])),
                        (min(moments3[:,0,2,1]), max(moments3[:,0,2,1]))),((min(moments3[:,1,0,1]), max(moments3[:,1,0,1])),
                        (min(moments3[:,1,1,1]), max(moments3[:,1,1,1])),(min(moments3[:,1,2,1]), max(moments3[:,1,2,1]))),
                        ((min(moments3[:,2,0,1]), max(moments3[:,2,0,1])),(min(moments3[:,2,1,1]), max(moments3[:,2,1,1])),
                        (min(moments3[:,2,2,1]), max(moments3[:,2,2,1])))) ])

    M3 = {'moments' : moments3,
          'mean' : M3mean,
          'range' : m3range,
          'st_dev' : M3st,
          'c_var' :  (M3st/M3mean) * 100
         }

    return M0, M1, M2, M3
```

```python
def ran_cells_M(n, m, l, s, g1):
    '''
    n - number of sets of particles
    m - size of set of particles
    l - loc parameter
    s - scale parameter
    g1 - lowesdst corner
    Maxwellian distribution
    '''
    cells = []

    for i in range(0,n):
        L = np.zeros((m, 7))
        L[:,0] = np.random.uniform(g1[0], g1[0] + 1.0, size=m).reshape(1,m)
        L[:,1] = np.random.uniform(g1[1], g1[1] +  1.0, size=m).reshape(1,m)
        L[:,2] = np.random.uniform(g1[2], g1[2] + 1.0, size=m).reshape(1,m)
        L[:,3] = np.random.normal(loc= l, scale= s, size=m).reshape(1,m)
        L[:,4] = np.random.normal(loc= l, scale= s, size=m).reshape(1,m)
        L[:,5] = np.random.normal(loc= l, scale= s, size=m).reshape(1,m)
        L[:,6] = np.random.uniform(0,  1.0, size=m).reshape(1,m)
        cells.append(L)

    return cells

def ran_cells_DM(n, m, l1, s1, l2, s2, g1):
    '''
    n - number of sets of particles
    m - size of set of particles
    l1 - loc parameter first maxwellian
    s1 - scale parameter first maxwellian
    l2 - loc parameter second maxwellian
    s2 - scale parameter second maxwellian
    g1 - lowesdst corner
    Double Maxwellian distribution
    '''
    cells = []

    for i in range(0,n):
        L = np.zeros((m, 7))
        L[:,0] = np.random.uniform(g1[0], g1[0] + 1.0, size=m).reshape(1,m)
        L[:,1] = np.random.uniform(g1[1], g1[1] +  1.0, size=m).reshape(1,m)
        L[:,2] = np.random.uniform(g1[2], g1[2] + 1.0, size=m).reshape(1,m)
        L[:,3] = (1/2)*(np.random.normal(loc= l1, scale= s1, size=m).reshape(1,m) + np.random.normal(loc= l2, scale= s2, size=m).reshape(1,m))
        L[:,4] = (1/2)*(np.random.normal(loc= l1, scale= s1, size=m).reshape(1,m) + np.random.normal(loc= l2, scale= s2, size=m).reshape(1,m))
        L[:,5] = (1/2)*(np.random.normal(loc= l1, scale= s1, size=m).reshape(1,m) + np.random.normal(loc= l2, scale= s2, size=m).reshape(1,m))
        L[:,6] = np.random.uniform(0,  1.0, size=m).reshape(1,m)
        cells.append(L)

    return cells

def kappa_pdf(k, v, w):
    f = lambda v : ((1 + v**2/(k * (w**2)))**(-(k+1)))
    N = inte.quad(f, -100, 100)[0]
    return (1/N)*((1 + v**2/(k * (w**2)))**(-(k+1)))

def rejection_sampling(n, D, R, pdf):
    '''
    n - number of observations required
    D - domain of distribution function
    R - range of distribution function
    pdf - probability distribution function
    '''
    observations = []
    lx, ux = D
    ly, uy = R
    counter = 0
    while len(observations) < n:
        x = np.random.uniform(lx, ux)
        y = np.random.uniform(ly, uy)
        if y < pdf(x):
            observations.append(x)

        counter = counter + 1
        if counter > 100*n**n:
            return print('Error')
    return observations

def ran_cells_K_rej(n, m, k, w, g1):
    '''
    n - number of sets of particles
    m - size of set of particles
    k - kappa parameter
    w - heat width
    g1 - lowest corner
    Kappa distribution
    '''
    cells = []

    R = (0, kappa_pdf(k, 0, w))
    D = (-0.1/k, 0.1/k)

    f = lambda v : ((1 + v**2/(k * (w**2)))**(-(k+1)))
    N = inte.quad(f, -100, 100)[0]

    def pdf(v):
        return (1/N)*((1 + v**2/(k * (w**2)))**(-(k+1)))

    for i in range(0,n):
        L = np.zeros((m, 7))
        L[:,0] = np.random.uniform(g1[0], g1[0] + 1.0, size=m).reshape(1,m)
        L[:,1] = np.random.uniform(g1[1], g1[1] +  1.0, size=m).reshape(1,m)
        L[:,2] = np.random.uniform(g1[2], g1[2] + 1.0, size=m).reshape(1,m)
        L[:,3] = rejection_sampling(m, D, R, pdf)
        L[:,4] = rejection_sampling(m, D, R, pdf)
        L[:,5] = rejection_sampling(m, D, R, pdf)
        L[:,6] = np.random.uniform(0, 1.0, size=m).reshape(1,m)
        cells.append(L)

    return cells

def ran_cells_3U(n, m, lx, ux, ly, uy, lz, uz, g1):
    '''
    n - number of sets of particles
    m - size of eqch set
    lx - lower bound x
    ux - upper bound x
    ly - lower bound y
    uy - upper bound y
    lz - lower bound z
    uz - upper bound z
    g1 - lowest corner
    3 unequal uniform distribution
    '''
    cells = []

    for i in range(0,n):
        L = np.zeros((m, 7))
        L[:,0] = np.random.uniform(g1[0], g1[0] + 1.0, size=m).reshape(1,m)
        L[:,1] = np.random.uniform(g1[1], g1[1] +  1.0, size=m).reshape(1,m)
        L[:,2] = np.random.uniform(g1[2], g1[2] + 1.0, size=m).reshape(1,m)
        L[:,3] = np.random.uniform(lx, ux, size=m).reshape(1,m)
        L[:,4] = np.random.uniform(ly, uy, size=m).reshape(1,m)
        L[:,5] = np.random.uniform(lz, uz, size=m).reshape(1,m)
        L[:,6] = np.random.uniform(0,  1.0, size=m).reshape(1,m)
        cells.append(L)

    return cells
```

```python
def continuous_moments(f, n):
    '''
    f - distribution function
    n - number of particles per cell (weights assumed to be from U(0,1)
    note :  space distribution assumed to be uniform
    '''
    moment0 = (inte.quad(f, -100, 100)[0]**3)*n/2

    fx = lambda v : f(v)*v

    fx2 = lambda v : f(v)*((v-(moment1_1/moment0))**2)

    fx21 = lambda v : f(v)*(v-(moment1_1/moment0))

    moment1_1 = (inte.quad(fx, -100, 100)[0]*(inte.quad(f, -100, 100)[0]**2))*n/2

    moment1 = np.zeros(3) + moment1_1

    moment2_2 = (inte.quad(fx2, -100, 100)[0]*(inte.quad(f, -100, 100)[0]**2))*n/2

    moment2_1 = ((inte.quad(fx21, -100, 100)[0]**2)*inte.quad(f, -100, 100)[0])*n/2

    moment2 = np.array([(moment2_2,moment2_1,moment2_1),(moment2_1,moment2_2,moment2_1),(moment2_1,moment2_1,moment2_2)])
    return moment0, moment1, moment2


def kappa_for_int(w,k):
    f = lambda v : ((1 + v**2/(k * (w**2)))**(-(k+1)))
    N = inte.quad(f, -100, 100)[0]
    return lambda v : (1/N) * ((1 + v**2/(k * (w**2)))**(-(k+1)))

def norm_for_int(mu, sig):
    return lambda x : (1/np.sqrt(2*np.pi*sig**2))*np.exp(-((x-mu)**2)/(2*sig**2))

def d_norm_for_int(mu1, sig1, mu2, sig2):
    return lambda x : (1/2)*((1/np.sqrt(2*np.pi*sig1**2))*np.exp(-((x-mu1)**2)/(2*sig1**2)) + (1/np.sqrt(2*np.pi*sig2**2))*np.exp(-((x-mu2)**2)/(2*sig2**2)))

f_k1 = kappa_for_int(1, 5)
f_n = norm_for_int(0, 1)
f_dn1 = d_norm_for_int(-0.5, 1, 0.5, 1)
f_dn2 = d_norm_for_int( 0, 0.1, 0, 3)


c_nor1_M0, c_nor1_M1, c_nor1_M2 = continuous_moments(f_n, 100)
c_dnor1_M0, c_dnor1_M1, c_dnor1_M2 = continuous_moments(f_dn1, 100)
c_dnor2_M0, c_dnor2_M1, c_dnor2_M2 = continuous_moments(f_dn2, 100)
c_kappa1_M0, c_kappa1_M1, c_kappa1_M2 = continuous_moments(f_k1, 100)


print(f"normal {c_nor1_M0, c_nor1_M1}")
print(f"{c_nor1_M2}")
print(f"d normal 1 {c_dnor1_M0, c_dnor1_M1}")
print(f"{c_dnor1_M2}")
print(f"d normal 2 {c_dnor2_M0, c_dnor2_M1}")
print(f"{c_dnor2_M2}")
print(f"Kappa {c_kappa1_M0, c_kappa1_M1}")
print(f"{c_kappa1_M2}")
```

```python
# implementing a first merging algorithm based on the above equations for 3->2 merging

def merger1(L, k, g1, pos):
    '''
    L - set of pqrticles to be merged
    k - desired number of particles
    g1 - smallest grid angle (i.e. the angle of the grid with lowest added coordinates value)
    '''
    L = list(L)
    s = len(L)

    while len(L) > k:
        p1 = np.zeros(7)
        p2 = np.zeros(7)
        p1_r = np.zeros(7)
        p2_r = np.zeros(7)
        score_r = 1000

        clus = clust_finder2(3, L)[0][0]

        w1 = clus[0,6]
        v1x = clus[0,3]
        v1y = clus[0,4]
        v1z = clus[0,5]
        w2 = clus[1,6]
        v2x = clus[1,3]
        v2y = clus[1,4]
        v2z = clus[1,5]
        w3 = clus[2,6]
        v3x = clus[2,3]
        v3y = clus[2,4]
        v3z = clus[2,5]

        aux = w1*v1x+w2*v2x+w3*v3x
        auy = w1*v1y+w2*v2y+w3*v3y
        auz = w1*v1z+w2*v2z+w3*v3z
        aPxx = w1*(v1x - aux)*(v1x - aux)+w2*(v2x - aux)*(v2x - aux)
        aPxy = w1*(v1x - aux)*(v1y - auy)+w2*(v2x - aux)*(v2y - auy)
        aPxz = w1*(v1x - aux)*(v1z - auz)+w2*(v2x - aux)*(v2z - auz)
        aPyz = w1*(v1z - auz)*(v1y - auy)+w2*(v2z - auz)*(v2y - auy)
        aPzz = w1*(v1z - auz)*(v1z - auz)+w2*(v2z - auz)*(v2z - auz)
        aPyy = w1*(v1y - auy)*(v1y - auy)+w2*(v2y - auy)*(v2y - auy)

        for sol in solutions:
            score = 0
            p1[6] = sol[0].subs(n,w1+w2)
            p2[6] = sol[1].subs(n,w1+w2)
            p1[3] = sol[2].subs([(n,w1+w2),(ux, aux), (Pxx, aPxx)])
            p1[4] = sol[3].subs([(n,w1+w2),(uy, auy), (Pyy, aPyy)])
            p1[5] = sol[4].subs([(n,w1+w2),(uz, auz), (Pzz, aPzz)])
            p2[3] = sol[5].subs([(n,w1+w2),(ux, aux), (Pxx, aPxx)])
            p2[4] = sol[6].subs([(n,w1+w2),(uy, auy), (Pyy, aPyy)])
            p2[5] = sol[7].subs([(n,w1+w2),(uz, auz), (Pzz, aPzz)])

            if abs(p1[3] + aux) != abs(p1[3]) + abs(aux):
                score = score + 1
                if abs(p1[3] + aux) >= abs(p1_r[3] + aux):
                    score = score + 1
            if abs(p1[4] + auy) != abs(p1[4]) + abs(auy):
                score = score + 1
                if abs(p1[4] + auy) >= abs(p1_r[4] + auy):
                    score = score + 1
            if abs(p1[5] + auz) != abs(p1[5]) + abs(auz):
                score = score + 1
                if abs(p1[6] + auz) >= abs(p1_r[6] + auz):
                    score = score + 1
            if abs(p2[3] + aux) != abs(p2[3]) + abs(aux):
                score = score + 1
                if abs(p2[3] + aux) >= abs(p2_r[3] + aux):
                    score = score + 1
            if abs(p2[4] + auy) != abs(p2[4]) + abs(auy):
                score = score + 1
                if abs(p2[4] + auy) >= abs(p2_r[4] + auy):
                    score = score + 1
            if abs(p2[5] + auz) != abs(p2[5]) + abs(auz):
                score = score + 1
                if abs(p2[6] + auz) >= abs(p2_r[6] + auz):
                    score = score + 1

            if score_r > score:
                p1_r = p1
                p2_r = p2
                score_r = score

        # methods for positioning of the particles
        if pos == 1: #first method, randomly following uniform distribution on the whole cell
            p1_r[0] = np.random.uniform(g1[0], g1[0] + 1.0)
            p1_r[1] = np.random.uniform(g1[1], g1[1] +  1.0)
            p1_r[2] = np.random.uniform(g1[2], g1[2] + 1.0)
            p2_r[0] = np.random.uniform(g1[0], g1[0] + 1.0)
            p2_r[1] = np.random.uniform(g1[1], g1[1] +  1.0)
            p2_r[2] = np.random.uniform(g1[2], g1[2] + 1.0)

        if pos == 2: #second method, randomly following uniform distribution within original particles limits
            p1_r[0] = np.random.uniform(min(clus[:,0]), max(clus[:,0]))
            p1_r[1] = np.random.uniform(min(clus[:,1]), max(clus[:,1]))
            p1_r[2] = np.random.uniform(min(clus[:,2]), max(clus[:,2]))
            p2_r[0] = np.random.uniform(min(clus[:,0]), max(clus[:,0]))
            p2_r[1] = np.random.uniform(min(clus[:,1]), max(clus[:,1]))
            p2_r[2] = np.random.uniform(min(clus[:,2]), max(clus[:,2]))

        if pos == 3: #third method, random uniform in extended particles limits
            xrange = min(max(clus[:,0]) - np.mean(clus[:,0]), np.mean(clus[:,0])- min(clus[:,0]))
            yrange = min(max(clus[:,1]) - np.mean(clus[:,1]), np.mean(clus[:,1])- min(clus[:,1]))
            zrange = min(max(clus[:,2]) - np.mean(clus[:,2]), np.mean(clus[:,2])- min(clus[:,2]))
            p1_r[0] = np.random.uniform(max(min(clus[:,0])-xrange/2, 1), min(max(clus[:,0])+xrange/2, 1))
            p1_r[1] = np.random.uniform(max(min(clus[:,1])-yrange/2, 1), min(max(clus[:,1])+yrange/2, 1))
            p1_r[2] = np.random.uniform(max(min(clus[:,2])-zrange/2, 1), min(max(clus[:,2])+zrange/2, 1))
            p2_r[0] = np.random.uniform(max(min(clus[:,0])-xrange/2, 1), min(max(clus[:,0])+xrange/2, 1))
            p2_r[1] = np.random.uniform(max(min(clus[:,1])-yrange/2, 1), min(max(clus[:,1])+yrange/2, 1))
            p2_r[2] = np.random.uniform(max(min(clus[:,2])-zrange/2, 1), min(max(clus[:,2])+zrange/2, 1))

        if pos == 4: #fourth method, randomly chosen position out of 2 of the original particles'
            first_pos = rand.randint(0,2)
            second_pos = first_pos
            while second_pos == first_pos:
                second_pos = rand.randint(0,2)
            if first_pos == 0:
                p1_r[0:3] = clus[0,0:3]
            elif first_pos == 1:
                p1_r[0:3] = clus[1,0:3]
            elif first_pos == 2:
                p1_r[0:3] = clus[2,0:3]
            if second_pos == 0:
                p2_r[0:3] = clus[0,0:3]
            elif second_pos == 1:
                p2_r[0:3] = clus[1,0:3]
            elif second_pos == 2:
                p2_r[0:3] = clus[2,0:3]

        if pos == 5: #fifth method, around weighted mean of original particles position
            avx = np.mean(max(clus[:,0])* clus[:,6])
            avy = np.mean(max(clus[:,1])* clus[:,6])
            avz = np.mean(max(clus[:,2])* clus[:,6])

            p1_r[0] = np.sum(clus[:,0]) - max(clus[:,0]) - min(clus[:,0])
            p1_r[1] = np.sum(clus[:,1]) - max(clus[:,1]) - min(clus[:,1])
            p1_r[2] = np.sum(clus[:,2]) - max(clus[:,2]) - min(clus[:,2])

            p2_r[0] = 2*avx-p1_r[0]
            p2_r[1] = 2*avy-p1_r[1]
            p2_r[2] = 2*avz-p1_r[2]

        if pos == 6: #sixth method, weighted random positions from previous particles positions
            w = np.sum(clus[:,6])
            poslist = [clus[0,0:3]] * int((clus[0,6]/w)*1000) + [clus[1,0:3]] * int((clus[1,6]/w)*1000)  + [clus[2,0:3]] * int((clus[2,6]/w)*1000)
            p1_r[0:3] = rand.choice(poslist)
            p2_r[0:3] = p1_r[0:3]
            while p2_r[0] == p1_r[0] and p2_r[1] == p1_r[1] and p2_r[2] == p1_r[2]:
                p2_r[0:3] = rand.choice(poslist)




        L = [el for el in L if not (np.array_equal(el, clus[0]) or np.array_equal(el, clus[1]) or np.array_equal(el, clus[2]))]
        L.append(p1_r)
        L.append(p2_r)

        #print(f'changes iteration {s - len(L)}')
        #print('before')
        #print(clus[0], clus[1], clus[2])
        #print('after')
        #print(p1_r, p2_r)



    return np.array(L)
```

```python
def continuous_moments_3uniform(n,lx,ux,ly,uy,lz,uz):
    '''
    f - istribution function
    n - number of particles per cell (weights assumed to be from U(0,1)
    note :  space distribution assumed to be uniform
    lx - lower bound x
    ux - uppper bound x
    ly - lower bound y
    uy - uppper bound y
    lz - lower bound z
    uz - uppper bound z
    '''
    fx = lambda v : 1/(ux - lx)
    fy = lambda v : 1/(uy - ly)
    fz = lambda v : 1/(uz - lz)

    moment0 = (inte.quad(fx, lx, ux)[0]*inte.quad(fy, ly, uy)[0]*inte.quad(fz, lz, uz)[0])*n/2

    fx1 = lambda v : fx(v)*v
    fy1 = lambda v : fy(v)*v
    fz1 = lambda v : fz(v)*v

    fxx2 = lambda v : fx(v)*(v-(moment1_x/moment0))**2
    fyy2 = lambda v : fy(v)*(v-(moment1_y/moment0))**2
    fzz2 = lambda v : fz(v)*(v-(moment1_z/moment0))**2
    fx2 = lambda v : fx(v)*(v-(moment1_x/moment0))
    fy2 = lambda v : fy(v)*(v-(moment1_y/moment0))
    fz2 = lambda v : fz(v)*(v-(moment1_z/moment0))

    moment1_x = (inte.quad(fx1, lx, ux)[0]*(inte.quad(fy, ly, uy)[0]*(inte.quad(fz, lz, uz)[0])))*n/2
    moment1_y = (inte.quad(fx, lx, ux)[0]*(inte.quad(fy1, ly, uy)[0]*(inte.quad(fz, lz, uz)[0])))*n/2
    moment1_z = (inte.quad(fx, lx, ux)[0]*(inte.quad(fy, ly, uy)[0]*(inte.quad(fz1, lz, uz)[0])))*n/2

    moment1 = np.array([moment1_x, moment1_y, moment1_z])

    moment2xx = (inte.quad(fxx2, lx, ux)[0]*(inte.quad(fy, ly, uy)[0]*(inte.quad(fz, lz, uz)[0])))*n/2
    moment2yy = (inte.quad(fyy2, ly, uy)[0]*(inte.quad(fx, lx, ux)[0]*(inte.quad(fz, lz, uz)[0])))*n/2
    moment2zz = (inte.quad(fzz2, lz, uz)[0]*(inte.quad(fx, lx, ux)[0]*(inte.quad(fz, lz, uz)[0])))*n/2

    moment2xy = (inte.quad(fx2, lx, ux)[0]*(inte.quad(fy2, ly, uy)[0]*(inte.quad(fz, lz, uz)[0])))*n/2
    moment2xz = (inte.quad(fx2, lx, ux)[0]*(inte.quad(fy, ly, uy)[0]*(inte.quad(fz2, lz, uz)[0])))*n/2
    moment2yz = (inte.quad(fx, lx, ux)[0]*(inte.quad(fy2, ly, uy)[0]*(inte.quad(fz2, lz, uz)[0])))*n/2
    moment2yx = moment2xy
    moment2zx = moment2xz
    moment2zy = moment2yz

    moment2 = np.array([(moment2xx,moment2xy,moment2xz),(moment2yx,moment2yy,moment2yz),(moment2zx,moment2zy,moment2zz)])

    return moment0, moment1, moment2
```
