```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
func = lambda x: 1/np.cosh(x)**2
```

```python
x = np.linspace(-4, 4)
y = func(x)
```

```python
plt.plot(x, y)
```

```python
z = np.gradient(y) # profile for B
```

```python
plt.plot(x, z)
```

```python
w = np.gradient(z) # profile for J
```

```python
plt.plot(x, w)
```
