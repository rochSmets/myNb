```python
import numpy as np
Na = 6.02*10**23
E  = 3.71*10**3
T  = 298
```

```python
def argPhi(dE) :
    return 1.5*Na*dE/E

def Phi(dE) :
    return np.exp(1.5*Na*dE/E)

def deltaS(dE) :
    return dE/T
```

```python
dE = 9.28*10**-27

print(100*(Phi(dE)-1))
print(deltaS(dE))
```

```python
dE = 6.24*10**-21

print(100*(Phi(dE)-1))
print(deltaS(dE))
```

```python
dE = 2.18*10**-18

print(100*(Phi(dE)-1))
print(deltaS(dE))
```
