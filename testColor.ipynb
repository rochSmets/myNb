```python
min_max = np.pi
N = 400
x = np.linspace(-min_max, min_max, N)
y = np.linspace(-min_max, min_max, N)
r = np.sqrt(x**2+y**2)
```

```python
xv, yv = np.meshgrid(x, y)
rv = 0.4*np.sqrt(xv**2+yv**2)
data = np.sin(xv)*np.sin(yv)*np.exp(-rv**2)

cmap = mpl.colormaps['viridis']
fig, ax = plt.subplots()
im = ax.imshow(data, cmap=cmap)
```
